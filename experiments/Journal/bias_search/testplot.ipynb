```python
# Import dependencies
import matplotlib.pyplot as plt
import pickle

with open('test.pkl', 'rb') as fp:
    traj = pickle.load(fp)

x=traj['x'][:,0]
y=traj['x'][:,1]

plt.plot(x,y, 'ro')

def connectpoints(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'k-')

for i in range(len(x)-1):
    connectpoints(x,y,i,i+1)

plt.axis('equal')
plt.show()
```
